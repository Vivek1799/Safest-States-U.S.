```python
import pandas as pd
#beautifulsoup4, write import statement here
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'https://www.consumeraffairs.com/homeowners/safest-states-in-the-us.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
table = soup.find('div', attrs={'class':'ftr-tb__wpr'})
```

```python
table
```

```python
# Find the table body
tbody = table.find('tbody')

# Find all rows in the table body
rows = tbody.find_all('tr')

# The list to hold dictionaries
data = []

# For each row, find all the 'td' tags and 'th' tag
for row in rows:
    cols = row.find_all('td')
    th = row.find('th')

    # Create a dictionary with data from each column
    row_data = {
        'State': th.text.strip(),
        'Murder score': cols[0].text.strip(),
        'Rape score': cols[1].text.strip(),
        'Robbery score': cols[2].text.strip(),
        'Aggravated assault score': cols[3].text.strip(),
        'Burglary score': cols[4].text.strip(),
        'Larceny-theft score': cols[5].text.strip(),
        'Motor vehicle theft score': cols[6].text.strip(),
        'Officers per 100k score': cols[7].text.strip(),
        'Final score': cols[8].text.strip(),
    }
    # Add the dictionary to the list
    data.append(row_data)

# Print the list of dictionaries
for item in data:
    print(item)
```

```python
import pandas as pd

# Assuming 'data' is the list of dictionaries obtained from the above script
df = pd.DataFrame(data)

# print the DataFrame
print(df)
```
